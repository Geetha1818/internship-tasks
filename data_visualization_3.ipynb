```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('EDA_Titanic_Task2(1).csv')
df.head()
```

```python
sns.countplot(data=df, x='Survived')
plt.title('Survival Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
print(df.columns)
```

```python
# Create Age groups column
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])

# Plot survival count by AgeGroup
sns.countplot(data=df, x='Survived', hue='AgeGroup')
plt.title('Survival Count by Age Group')
plt.show()
```

```python
sns.countplot(data=df, x='Survived')
plt.title('Overall Survival Count')
plt.show()
```

```python
plt.hist(df['Age'].dropna(), bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.scatter(df['Age'], df['Survived'])
plt.title('Survived vs Age')
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show()
```
